```python
#problem3
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Create a undirected graph
A = nx.DiGraph()

# Add nodes
A.add_node(1)
A.add_node(2)
A.add_node(3)
A.add_node(4)
A.add_node(5)

edges1 = [(1, 2), (2, 3), (3, 4), (4,5)]
A.add_edges_from(edges1)

adj_matrix = nx.adjacency_matrix(A)
adj_array = adj_matrix.toarray()


# Compute the degree matrix
degree_matrix = np.diag(np.sum(adj_array, axis=1))

# Compute the Laplacian matrix
laplacian_matrix = degree_matrix - adj_array
# Compute the eigenvalues
eigenvalues = np.linalg.eigvals(laplacian_matrix)

# Set epsilon to 0.8
epsilon = 0.8

# Compute I + epsilon * L
I_plus_epsilon_L = np.eye(laplacian_matrix.shape[0]) + epsilon * laplacian_matrix

# Compute eigenvalues of I + epsilon * L
eigenvalues = np.linalg.eigvals(I_plus_epsilon_L)

# Check stability: All eigenvalues must have negative real parts
if np.all(np.real(eigenvalues) > 0):
    # Check consensus: Look for single eigenvector associated with eigenvalue 0
    # If found, consensus is achieved, determine consensus value
    zero_index = np.where(np.isclose(eigenvalues, 0))[0]
    if len(zero_index) == 1:
        eigenvector_zero = np.real(np.linalg.eig(I_plus_epsilon_L)[1][:, zero_index[0]])
        consensus_value = np.mean(eigenvector_zero)
        print(f"Consensus achieved with epsilon={epsilon}. Consensus value: {consensus_value}")
    else:
        print(f"System is stable but consensus cannot be achieved with epsilon={epsilon}")
else:
    print(f"System is unstable for epsilon={epsilon}")

'''
print("Laplacian Matrix:")
print(laplacian_matrix)
print("\nAdjacency Matrix:")
print(adj_array)
print("\nDegree Matrix:")
print(degree_matrix)
print("\neigenValues")
print(eigenvalues)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(A)  # positions for all nodes
nx.draw(A, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
plt.title("Original Graph")
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle

T = 0.01
Timesteps = 1000
N = 3  # Number of robots
d = 1  # Distance parameter
A = np.ones((N, N))  # Adjacency matrix (all ones for now)

# Initial conditions
x_initial = [2, 1, 3]  # x coordinates
y_initial = [1, 2, 3]  # y coordinates
theta_initial = [0, -np.pi/2, np.pi/2]  # orientations

# Leader communication vector
C_leader = [1, 1, 1]

# Initial values of robot configurations
x = np.zeros((Timesteps + 1, N))
y = np.zeros((Timesteps + 1, N))
theta = np.zeros((Timesteps + 1, N))

x[0] = x_initial
y[0] = y_initial
theta[0] = theta_initial

x_leader = np.zeros(Timesteps)
y_leader = np.zeros(Timesteps)

for k in range(Timesteps):
    # Trajectory for leader
    x_leader[k] = 3 * np.cos(2 * np.pi * 0.5 * k * T)
    y_leader[k] = 3 * np.sin(2 * np.pi * 0.5 * k * T)

    for i in range(N):
        z1 = x[k, i] + d * np.cos(theta[k, i])
        z2 = y[k, i] + d * np.sin(theta[k, i])

        V1 = 0
        V2 = 0
        for j in range(N):
            V1 += A[i, j] * (z1 - z1) + C_leader[i] * (x_leader[k] - z1)
            V2 += A[i, j] * (z2 - z2) + C_leader[i] * (y_leader[k] - z2)

        YY = np.linalg.inv([[np.cos(theta[k, i]), -d * np.sin(theta[k, i])],
                            [np.sin(theta[k, i]), d * np.cos(theta[k, i])]]) @ np.array([V1, V2])
        v = YY[0]
        omega = YY[1]

        x[k + 1, i] = x[k, i] + T * v * np.cos(theta[k, i])
        y[k + 1, i] = y[k, i] + T * v * np.sin(theta[k, i])
        theta[k + 1, i] = theta[k, i] + T * omega

# Calculate the range of consensus for x and y values
x_consensus_range = []
y_consensus_range = []

for i in range(N):
    x_consensus_range.append(np.max(x[:, i]) - np.min(x[:, i]))
    y_consensus_range.append(np.max(y[:, i]) - np.min(y[:, i]))

# Print the range of consensus for x and y values for each robot
for i in range(N):
    print(f"Robot {i+1}:")
    print(f"  Range of consensus for x: {x_consensus_range[i]}")
    print(f"  Range of consensus for y: {y_consensus_range[i]}")

# Plotting paths
plt.figure()
plt.plot(x_leader, y_leader, 'r', label='Leader')
for i in range(N):
    plt.plot(x[:, i], y[:, i], label=f'Robot {i+1}')
    plt.gca().add_patch(Circle((x[0, i], y[0, i]), 0.2, color='blue', alpha=0.5))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Paths of Robots')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
