```python
#problem1
import numpy as np
import matplotlib.pyplot as plt

#this is the cost function
def g(w):
    return np.log(1 + np.exp(w**2))
    
#this is the first derivative that we will use for gradient descent
def g_1p(w):
    return (2*np.exp(w**2) * w) / (1+ np.exp(w**2))

def gradientDescent(w_init,learningRate,k):
    w_values = [w_init]
    cost_values = [g(w_init)]

    for _ in range(k):
        w_new = w_values[-1] - learningRate * g_1p(w_values[-1])
        w_values.append(w_new)
        cost_values.append(g(w_new))

    return w_values, cost_values

#function that graphs other functions (ONLY WORKS WITH BASIC FUNCTIONS)
def graphFunction(func, k):
    function = np.zeros(k)
    for i in range(k):
        function[i] = func(i)
    plt.plot(function)
    

k=100  #number of iterations

w_values, cost_values = gradientDescent(1, 0.1, k)   # Run gradient descent

#adjust sizing
plt.figure(figsize=(10, 6))

# Plot w(k) curve
plt.subplot(1, 2, 1)
plt.plot(w_values, label='w(k)')
plt.title('Gradient Descent: w(k)')
plt.xlabel('Iterations')
plt.ylabel('w(k)')
plt.legend()

# Plot g(w(k)) curve
plt.subplot(1, 2, 2)
plt.plot(cost_values, label='g(w(k))', color='orange')
plt.title('Gradient Descent: g(w(k))')
plt.xlabel('Iterations')
plt.ylabel('g(w(k))')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#problem2
import numpy as np
import matplotlib.pyplot as plt

#this is the cost function
def g(w):
    return np.log(1 + np.exp(w**2))
    
#this is the first derivative that we will use for gradient descent
def g_1p(w):
    return (2*np.exp(w**2) * w) / (1+ np.exp(w**2))
    
#this is the second derivative that we will use for the newton method of approximation
def g_2p(w):
    max_w = np.max(w)
    exp_term = np.exp(w - max_w)
    numerator = 2 * exp_term * (2 * w**2 + exp_term + 1)
    denominator = (1 + exp_term)**2
    result = numerator / denominator
    return result


#function that graphs other functions (ONLY WORKS WITH BASIC FUNCTIONS)
def graphFunction(func, k):
    function = np.zeros(k)
    for i in range(k):
        function[i] = func(i)
    plt.plot(function)

#graphFunction(g_2p,1000)

def newtonApproximation(w_init,g_1p,g_2p,k):
    w = np.zeros(k)
    w[0] = w_init
    
    for i in range(1, k):
        w[i] = w[i-1] - (g_1p(w[i-1])) / (g_2p(w[i-1]))
    return w

w_init = 1
k=100
w = newtonApproximation(w_init,g_1p,g_2p,k)

# Calculate g(w(k)) for each iteration
g_values = [g(w) for w in w]

# Plot the Newton approximation
plt.figure(figsize=(8, 6))
plt.plot(g_values, marker='o', label='g(w(k))', color='orange')
plt.title("Newtons Method: g(w(k))")
plt.xlabel('Iterations')
plt.ylabel('w')
plt.legend()
plt.show()
```
