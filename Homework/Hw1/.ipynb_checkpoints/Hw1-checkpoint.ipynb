```python
import numpy as np
import matplotlib.pyplot as plt

def vicsek_model(num_particles, time_steps, V, ni):
    # Initialize positions and headings randomly
    positions = np.random.rand(num_particles, 2)
    headings = np.random.rand(num_particles) * 2 * np.pi

    for k in range(time_steps):
        # Update headings
        headings += np.sum(headings) / num_particles + 1 / (1 + ni) * np.random.uniform(-np.pi, np.pi, num_particles)

        # Update positions
        positions[:, 0] += V * np.cos(headings)
        positions[:, 1] += V * np.sin(headings)

        # Plot current positions
        plt.scatter(positions[:, 0], positions[:, 1], c='blue', marker='o', s=10)

    plt.title('Vicsek Model Simulation')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Example usage with 100 particles over 100 time steps
vicsek_model(num_particles=100, time_steps=100, V=0.1, ni=0.1)
```
